```python
# Smart Parking Lot Management System
def smart_parking_system(capacity, logs):
    current_vehicles = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT" and current_vehicles > 0:
            current_vehicles -= 1

        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    if current_vehicles > capacity:
        status = "Capacity Exceeded"
    elif current_vehicles == capacity:
        status = "Full"
    else:
        status = "Available"

    return current_vehicles, peak_usage, status

capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

current, peak, status = smart_parking_system(capacity, logs)
print("Currently Parked Vehicles:", current)
print("Peak Usage:", peak)
print("Parking Status:", status)
```

```python
# Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5

    if traffic == "High":
        base_time += 10
    elif traffic == "Medium":
        base_time += 5

    if weather == "Rainy":
        base_time += 5
    elif weather == "Storm":
        base_time += 15

    return base_time


eta = estimate_delivery_time(8, "High", "Rainy")
print("Estimated Delivery Time:", eta, "minutes")
```

```python
# Movie Theatre Seat Occupancy
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100

    occupancy_percentage = int(occupancy_percentage)

    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    return occupancy_percentage, status

total_seats = 200
booked_seats = [1] * 150

occupancy, status = theatre_occupancy(total_seats, booked_seats)
print("Occupancy:", str(occupancy) + "%")
print("Show Status:", status)
```

```python
# Cloud Server Load Classification System
def server_load_classifier(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)
    average_cpu = int(average_cpu)

    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return average_cpu, status


readings = [45, 60, 70, 85, 90]
avg_cpu, status = server_load_classifier(readings)
print("Average CPU Load:", str(avg_cpu) + "%")
print("Server Status:", status)
```

```python
# Smart Classroom Resource Usage Monitor
def classroom_resource_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        alert = "Yes"
        overused_output = ", ".join(overused)
    else:
        alert = "No"
        overused_output = "None"

    return overused_output, alert

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused_resources, energy_alert = classroom_resource_monitor(usage)

print("Overused Resources:", overused_resources)
print("Energy Alert:", energy_alert)
```

```python
# Online Event Registration Capacity Controller
def event_registration_controller(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    return confirmed, waitlisted, status

capacity = 100
registrations = 105

confirmed, waitlisted, status = event_registration_controller(capacity, registrations)
print("Confirmed Registrations:", confirmed)
print("Waitlisted Users:", waitlisted)
print("Registration Status:", status)
```
