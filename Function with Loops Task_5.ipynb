```python
# Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status

likes = [200, 300, 150, 400]
total, result = analyze_post_engagement(likes)
print("Total Likes:", total)
print("Post Status:", result)
```

```python
# Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status

rainfall = [70, 75, 80, 63]
avg, result = rainfall_checker(rainfall, 70)
print("Average Rainfall:", avg)
print("Rainfall Status:", result)
```

```python
# Social Media – Duplicate Account Detection
def detect_duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"

accounts = ["john123", "alice99", "john123", "mark22"]
print("Duplicate Accounts Found:", detect_duplicate_accounts(accounts))
```

```python
# System Monitoring – Application Health Checker
def application_health_checker(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return status

errors = 7
print("Error Count:", errors)
print("System Status:", application_health_checker(errors))
```

```python
# E-Learning – Student Attendance Eligibility System
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for day in attendance_list:
        if day == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return percentage, status

attendance = [1,1,1,0,1,1,1,1,0,1]
percent, result = attendance_eligibility(attendance)
print("Attendance Percentage:", percent)
print("Exam Eligibility:", result)
```

```python
# Smart Electricity Bill Analyzer
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

units = 250
total_bill, usage_status = electricity_bill(units)
print("Total Bill:", total_bill)
print("Usage Classification:", usage_status)
```
