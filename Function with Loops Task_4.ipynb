```python
#Mobile Recharge Validation System
def mobile_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Invalid! Recharge amount must be at least ₹50.")
            continue

        if amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid plan! Please choose a valid plan:", valid_plans)


mobile_recharge()
```

```python
#Inventory Reorder Alert System
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 12,
    "Milk": 30
}

check_inventory(inventory)
```

```python
#Student Result Processing System
def student_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks_list = [60, 70, 45, 55, 80]

result = student_result(marks_list)
print("Result:", result)
```

```python
#Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare *= 1.25   # 25% extra

    return fare


while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(km, peak_hour)

        print(f"Total Fare: ₹{total_fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break

    except ValueError:
        print("Please enter valid input!")
```

```python
#Employee Attendance Eligibility Checker
def attendance_checker(attendance):
    present_days = 0

    for day in attendance:
        if day.upper() == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    print(f"Attendance Percentage: {percentage:.2f}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "A", "P", "P", "A", "P", "P"]

status = attendance_checker(attendance_list)
print("Status:", status)
```

```python
#Password Strength Checker
def password_checker(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password  (Minimum length is 8)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password(Must include a digit and special character)"


user_password = input("Enter your password: ")
print(password_checker(user_password))
```
